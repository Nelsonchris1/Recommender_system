```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from math import sqrt
%matplotlib inline
print('Setup complete')
```

```python
movies_df = pd.read_csv('movies.csv')
rating_df = pd.read_csv('ratings.csv')

movies_df.head()
```

```python
movies_df.shape
```

```python
#Remove year from title and store in a new column year
#USing pandas regualr expression to find year stored between parenthesis
#we specify the parenthesis so we dont cinflict with movies that have years in their titles

movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand = False)

#Next we remove the parenthesis

movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand  = False)

#Now remove the year from title column
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')

#Applying strip to remove and whitespace created in the title column
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
```

```python
movies_df.head()
```

```python
#The genre column is separated by | so we call the split function on |
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
movies_df.tail(10)
```

```python
movies_df.isnull().sum()
```

```python
moviesGenres_df = movies_df.copy()

#For every row in the dataset iterate through list of genre and assign 1 into the corresponding columns
for index,row in movies_df.iterrows():
    for genre in row['genres']:
        moviesGenres_df.at[index,genre] = 1

        
#Filling NaN with 0 to show that mmovie doesnt have that genre
moviesGenres_df = moviesGenres_df.fillna(0)
moviesGenres_df.head()
```

```python
rating_df.shape
```

```python
rating_df.drop(columns = 'timestamp',inplace = True)
```

## Content based recommendation

```python
#create a input user to recommend movie to
User_input = [
                {'title': 'Breakfast Club, The' , 'rating' : 5},
                {'title': 'Toy Story', 'rating': 3.5},
                {'title': 'Jumanji', 'rating' : 2},
                {'title': 'Pulp Fiction', 'rating': 5},
                {'title': 'Akira','rating': 4.5},
                {'title': 'Silver Spoon', 'rating': 3.5}
             ]

input_movies_df = pd.DataFrame(User_input)
input_movies_df
```

```python
#extract the movie ID of our user input from movie column

#filtering out the movies by title
inputid = movies_df[movies_df['title'].isin(input_movies_df['title'].tolist())]

#Then merging it so we can get the movieId. It is implicitly merging it by title.
input_movies_df = pd.merge(inputid,input_movies_df)
```

```python
input_movies_df
```

```python
#Drop informations we wont use from the input dataframe
input_movies_df.drop(columns = ['genres','year'], inplace = True)
```

```python
input_movies_df
```

```python
#To get the users prefrence where we include genres

userMovies = movies_df[movies_df['movieId'].isin(input_movies_df['movieId']).tolist()]
userMovies
```

```python
userMovies = moviesGenres_df[moviesGenres_df['movieId'].isin(input_movies_df['movieId'].tolist())]
userMovies
```

```python
userMovies
```

```python
#Reset index of the userMovie dataframe and drop movieId,titlee,genre and year

userMovies = userMovies.reset_index(drop = True)

#drop columns

user_genre = userMovies.drop (columns = ['movieId','title','genres','year'])
user_genre
```

```python
input_movies_df['rating']
```

```python
#Perform a dot matrix to get the weight on the rating in user_genre
    
user_profile = user_genre.transpose().dot(input_movies_df['rating'])
user_profile
```

```python
#get the genre of everymovie from our original dataset
genreTable = moviesGenres_df.set_index(moviesGenres_df['movieId'])

#drop some columns
genreTable.drop(columns = ['title','movieId','year','genres'],inplace = True)
genreTable.head()
```

```python
genreTable.shape
```

```python
'''
with the inputs profile and the complete list of movies and their genres in hand,
we're going to take the wrighted average of every movie based
on their input profile and recommend the top twenty movies that most satisfy it
'''

#Multiply the genreTable by the usersrofile and then average it by the sum of the usersprofile

RecommendationTable_df = ((genreTable*user_profile).sum(axis = 1))/ user_profile.sum()
RecommendationTable_df.head()
```

```python
# to sort our recommendation table
RecommendationTable_df = RecommendationTable_df.sort_values(ascending = False)
RecommendationTable_df.head(10)
```

```python
#Final Recommender system table
movies_df.loc[movies_df['movieId'].isin(RecommendationTable_df.head(10).keys())]
```
